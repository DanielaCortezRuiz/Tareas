Funcion

Una funci√≥n es un conjunto de instrucciones que se agrupan para realizar una tarea espec√≠fica.

Recibe datos (entrada), los procesa y devuelve un resultado (salida).
Sirven para reutilizar c√≥digo, organizar mejor los programas y evitar repetir instrucciones.


def:	palabra clave para definir una funci√≥n

Par√°metros:	datos que la funci√≥n recibe

return:	indica qu√© valor devuelve la funci√≥n

Llamada:	cuando usamos la funci√≥n en el programa

```python
def sumar(a, b):  #La funci√≥n recibe dos n√∫meros, realiza una operaci√≥n y devuelve el resultado.
    resultado = a + b
    return resultado
```

```python
sumar(10,25) #Llamar la funci√≥n suma
```

```python
def resta(a,b):
    res = a - b
    return res
```

```python
resta(10, 25)
```

Ejercicio: Crear una funci√≥n que convierta temperatura Celsius a Kelvin

```python
def conversionC_K(T_C):
    res= T_C + 273.15
    return res
```

```python
conversionC_K(30)
```

Ejercicio: Funci√≥n que cambie de C a K, F a K

```python
def a_K(T, unidad):
    if unidad == "C":
       res= T +273.15
       return res
    elif unidad == "F":
      res = (T - 32)* 5/9 +273.15
      return res
    elif unidad == "R":
       res = T* 5/9
       return res
    else:
      print("Dato no v√°lido, verifique las unidades")
```

```python
a_K(30, "F")
```
